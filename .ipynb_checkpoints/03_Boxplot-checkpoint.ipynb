```python
# Importing the required module
import seaborn as sns

# Set style of background of plot
sns.set(style="whitegrid")

# Load dataset
tip = sns.load_dataset("tips")

# Define a color palette for 'smoker' categories
custom_palette = {"Yes": "#d5e809", "No": "#8c30ba"}

# Create a box plot with individual colors for each hue level
sns.boxplot(x="tip", y="day", hue="smoker", data=tip, 
            palette=custom_palette, dodge=True)
```

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

kashit =sns.load_dataset('titanic')
sns.boxplot(x='survived',
            y='age',
           data=kashit)
plt.show
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

sns.boxplot(x='survived',
            y='age', showmeans=True,
            meanprops={"marker": "*",
                      "markersize":"12",
                       "markeredgecolor": "black"},
           data= kashit)
#show labels
plt.xlable:("how many survived"),
plt.ylable("Age (years)"),
plt.title("box plot of kitnay doobay and kitnay bach gaye")

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set up to handle Unicode in the title
plt.rcParams["axes.unicode_minus"] = False

sns.boxplot(x='survived',
            y='age', showmeans=True,
            meanprops={"marker": "*",
                       "markersize": 12,
                       "markeredgecolor": "black"},
            data=kashit)

# Show labels
plt.xlabel("Survived")
plt.ylabel("Age (years)")
plt.title("Box Plot of Survival by Age")

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
kashti = sns.load_dataset("titanic")
sns.boxplot(x="survived",
             y="age",
             showmeans=True,
             meanprops= {"marker":"+",
                         "markersize":"21",
                         "markercolor":"red"},
             data = Kashti)
plt.xlabel("How Many survived"),
plt.ylabel("Agen (In Years)"),
plt.title("Box plot of kitnay doobay and kitnay bach gaye")

plt.show()
```
